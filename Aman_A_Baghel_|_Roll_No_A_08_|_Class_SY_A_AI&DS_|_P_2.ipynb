Q1. What is data ingestion and why is it significant in data science? Which Python library
is most commonly used?  
Data ingestion refers to the process of gathering, importing, and loading data from diverse
sources — including CSV files, databases, REST APIs, and web services — into a
processing environment for further analysis. It forms the entry point of any data science
pipeline, and the quality of ingested data directly determines the reliability of downstream
models and insights. Without structured ingestion, inconsistencies and errors can propagate
undetected. In Python, Pandas is the most widely adopted library for this purpose, o ering
powerful DataFrame structures and read functions compatible with CSV, Excel, JSON, and
SQL formats.  
Q2. Why must data be cleaned before analysis or model training?  
Real-world datasets frequently contain imperfections such as missing entries, repeated
records, inconsistent formatting, and outlier values. If left unaddressed, these anomalies can
distort statistical summaries and lead machine learning models to learn incorrect patterns,
reducing predictive accuracy. Data cleaning resolves these issues by standardizing data
types, filling or removing null values, eliminating duplicates, and correcting format
inconsistencies. This ensures that the dataset accurately represents the real-world
phenomenon being studied, resulting in more trustworthy analysis and better-performing
models.  
Q3. How is a CSV file loaded into a Pandas DataFrame?  
Loading a CSV file into a Pandas DataFrame is accomplished using the pd.read_csv()
function. This method accepts the file path as an argument and returns a structured two
dimensional DataFrame object. Users can customize the behavior with optional parameters
to handle headers, specify delimiters, manage encoding, and parse dates. The function is
robust and widely used due to its simplicity and ability to handle large files with minimal
code.  
Q4. What is the di
erence between the head() and tail() functions?  
The head() function in Pandas returns the first n rows (default: 5) of a DataFrame, making
it useful for inspecting the beginning of a dataset and verifying that data loaded correctly.
Conversely, tail() returns the last n rows, helpful for confirming the end of the dataset and
spotting any trailing anomalies. Both functions support a numeric argument to control how
many rows are displayed and are particularly useful when working with large files where
displaying the full dataset would be impractical.  
Q5. How are missing values identified in a Pandas DataFrame?  
Missing values can be identified using isnull() or isna(), both of which return a boolean
mask indicating positions where values are absent. Chaining with .sum() produces a
column-wise count of missing entries, providing a quick overview of data completeness.
The info() method also shows non-null counts per column. Together, these tools allow
analysts to assess the extent of missing data before deciding on an appropriate handling
strategy — whether that means imputation, row removal, or leaving them as-is based on
context.  

```python
# Download the titanic.csv dataset if it's not already present
!mkdir -p data
!wget -nc https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv -P data

# Now the file should be located at 'data/titanic.csv'
```

```python
# Implementation
# Loading the Dataset
import pandas as pd
import numpy as np

# Load CSV from uploaded file
df = pd.read_csv('data/titanic.csv')
df.head()

# Exploring Dataset Structure
# Display schema and data types
df.info()

# Summary statistics for numerical columns
df.describe()

# Identifying Missing Values
# Count of null values per column
print(df.isnull().sum())

# Handling Missing Values
# Replace all missing values with zero
df = df.fillna(0)
print('Missing values after fillna:', df.isnull().sum().sum())

# Column Name Standardization
# Convert column headers to lowercase
df.columns = df.columns.str.lower()
df.head()

# Dataset Dimensions
print('Shape (rows, columns):', df.shape)

# Removing Duplicate Records
df = df.drop_duplicates()
print('Shape after deduplication:', df.shape)

# Imputing with Median
# Fill remaining numerical nulls using median
df.fillna(df.median(numeric_only=True), inplace=True)

# Data Type Conversion
df['age'] = df['age'].astype(int)
print(df.dtypes)

# Filtering Rows by Condition
# Passengers older than 30
senior = df[df['age'] > 30]
print(senior.head())

# Saving Cleaned Data
df.to_csv('cleaned_titanic.csv', index=False)
print('Cleaned dataset saved successfully.')
```
