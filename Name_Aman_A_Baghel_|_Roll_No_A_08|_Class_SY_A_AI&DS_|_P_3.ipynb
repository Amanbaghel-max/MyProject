Q1. What is Data Wrangling? How does it di
er from Data Preprocessing?  
Data wrangling, also called data munging, is the comprehensive process of converting raw,
disorganized data into a clean and structured format ready for analysis. It includes
collecting data from multiple sources, handling missing or inconsistent entries,
restructuring the layout, converting data types, and merging datasets. Data preprocessing,
while related, is a more narrowly scoped activity that occurs after wrangling, primarily in
machine learning pipelines. Preprocessing involves modelspecific transformations such as
feature scaling, normalization, label encoding, and train-test splitting — activities aimed at
preparing data for algorithmic consumption rather than general analytical readability.  
Q2. What role does Exploratory Data Analysis (EDA) play when using Pandas?  
Exploratory Data Analysis is an analytical approach that uses visual and statistical methods
to understand a dataset's core structure before applying formal modeling. In Pandas, EDA is
typically initiated using head(), which reveals the first few rows; info(), which summarizes
column types and null counts; and describe(), which presents key statistical metrics like
mean, standard deviation, and percentiles. EDA helps analysts identify outliers, detect data
imbalances, understand distributions, and discover relationships between variables —
forming the foundation for any subsequent analysis or model building.

Q4. Explain the purpose of isnull(), notnull(), dropna(), and fillna() in Pandas.  
These four functions collectively manage missing data in a DataFrame. isnull() returns a
boolean mask where True marks absent values, allowing detection of gaps in the data.
notnull() is its complement, returning True where values are present. dropna() removes
rows or columns containing null values — appropriate when the proportion of missing data
is small and deletion won't significantly bias the dataset. fillna() replaces null entries with a
user-specified substitute such as the column mean, median, a constant, or a forward-filled
value, preserving the row count while restoring completeness.  
Q5. What is column transformation and why is it important?  
Column transformation involves modifying existing columns or deriving new ones from
existing data to improve the dataset's usefulness for analysis. Common transformations
include arithmetic calculations to create derived features, string operations for formatting,
data type casting, categorical encoding, and normalization. These transformations enrich
the dataset, ensure consistency, and make features suitable for machine learning algorithms.
For example, multiplying sepal length and width to produce a sepal area provides a
composite feature that may carry more predictive power than either dimension alone.

```python
import pandas as pd
import numpy as np
records = {
'Name': ['Jiya', 'Riya', 'Tiya', 'Ram', 'Jay'],
'Age':   [20, 21, np.nan, 23, 22],
'Marks': [85, np.nan, 78, 92, 88]
}
df = pd.DataFrame(records)
print('Original Dataset:')
print(df)
# Exploratory Data Analysis
print(df.head())
# First 5 rows
print(df.info())
# Schema summary
print(df.describe())    # Statistical overview
# Missing Value Detection
print('Null Mask:')
print(df.isnull())
print('Null Counts:')
print(df.isnull().sum())
print('Non-null Mask:')
print(df.notnull())
# Removing Rows with Nulls
df_no_null = df.dropna()
print('After dropna():')
print(df_no_null)
# Filling Missing Values with Mean
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
print('After fillna():')
print(df)
# Column Transformation
# Derive a new column
df['Total'] = df['Marks'] + 10 # Convert float age to integer
df['Age'] = df['Age'].astype(int)
print('Transformed Dataset:')
print(df)
# Wrangling on the Iris Dataset
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
iris.head()
# Iris EDA and Null Check
iris.info()
iris.describe()
print(iris.isnull().sum())
# Introducing and Handling Missing Values
iris.loc[0, 'sepal_length'] = np.nan
iris.loc[1, 'petal_width'] = np.nan
iris['sepal_length'].fillna(iris['sepal_length'].mean(), inplace=True)
iris['petal_width'].fillna(iris['petal_width'].mean(), inplace=True)
print(iris.isnull().sum())
# Feature Engineering and Aggregation
iris['sepal_area'] = iris['sepal_length'] * iris['sepal_width']
print(iris.groupby('species').mean())
print(iris.sort_values(by='sepal_length', ascending=False).head())
```
