Q1. What are sequence data types in Python? Explain indexing and slicing with examples.  
In Python, sequence data types are ordered collections that allow access to individual
elements through their position (index). The primary sequence types are strings, lists,
tuples, arrays, and range objects. Indexing begins at 0 for the first element and Python also
supports negative indexing where -1 points to the last element. Slicing is a technique to
extract a portion of a sequence using the syntax sequence[start:end:step], which provides a
subset without altering the original. These features are consistent across all sequence types,
making Python a versatile language for data handling.  
Q2. What distinguishes a list from a tuple in Python? When should tuples be preferred?  
Both lists and tuples are ordered collections in Python, but their fundamental di
erence lies in mutability. Lists, enclosed in square brackets [], are mutable — meaning
elements can be added, modified, or removed after creation. Tuples, enclosed in
parentheses (), are immutable — once defined, their contents cannot be changed. This
immutability makes tuples faster to process and more memory-e icient than lists. Tuples are
preferred when storing data that should remain fixed throughout a program's execution,
such as geographic coordinates, configuration values, or database row records, where
accidental modification would be problematic.  
Q3. How do Python arrays di
er from lists, and when is it appropriate to use arrays?  
While both arrays and lists store sequences of items, Python's array module enforces type
homogeneity — every element must share the same data type. This constraint makes arrays
more memory-e
icient and faster for numerical processing compared to lists, which
can hold mixed types. Arrays are the preferred structure when dealing with large datasets of
uniform numeric values, scientific computations, or situations where memory e
iciency is critical. For general-purpose usage with mixed types, lists remain more versatile.  
Q4. Explain how Python dictionaries work internally. What makes them e
icient?  
Python dictionaries store data as key-value pairs and use a hash table as their underlying
data structure. When a key is provided, Python applies a hash function to compute a unique
memory location where the associated value is stored. This mechanism enables O(1)
average-time complexity for lookups, inserts, and deletions. By contrast, searching through
a list or tuple requires O(n) time in the worst case. This makes dictionaries exceptionally e
icient for applications requiring rapid data retrieval, such as caching, frequency counting,
and symbol tables.  
Q5. How do sets di
er from lists in Python, and how does a set handle duplicate values?  
Sets are unordered collections that enforce uniqueness among their elements, whereas lists
maintain insertion order and permit duplicates. When a duplicate value is added to a set, it
is silently ignored, ensuring all members remain distinct. Sets are optimized for
membership testing and support mathematical operations like union, intersection, di erence,
and symmetric di
erence. These characteristics make sets ideal for tasks such as
deduplication of records, common element detection, and e
icient membership checks
that would otherwise require looping through a list.  
Implementation  
String Operations



```python
s = "Hello Python"
print('String:', s)
print('Index [0]:', s[0])          # First character
print('Index [-1]:', s[-1])         # Last character
print('Slice [0:5]:', s[0:5])       # First word
print('Slice [6:]:', s[6:])         # Second word
print('Reversed:', s[::-1])         # Reverse the string

# List Operations
numbers = [10, 20, 30, 40, 50]
print('List:', numbers)
print('Element at index 0:', numbers[0])
print('Slice [1:4]:', numbers[1:4])
numbers.append(60)      # Add element at end
numbers.remove(20)      # Remove specific element
print('After modification:', numbers)

# Array Operations
import array

int_arr = array.array('i', [1, 2, 3, 4, 5])
print('Array:', int_arr)
print('First element:', int_arr[0])
print('Slice [1:4]:', int_arr[1:4])
int_arr.append(6)
print('After append:', int_arr)

# Tuple Operations
coords = (100, 200, 300, 400)
print('Tuple:', coords)
print('First element:', coords[0])
print('Slice [1:3]:', coords[1:3])
# Tuples are immutable — no modification allowed

# Dictionary Operations
student = {'name': 'Arjun', 'age': 20, 'course': 'BTech'}
print('Dictionary:', student)
print('Name:', student['name'])
student['city'] = 'Pune'     # Add new key
student['age'] = 21          # Update existing key
print('Updated Dict:', student)

# Set Operations
unique_nums = {1, 2, 3, 4, 4, 5}    # Duplicate 4 auto-removed
print('Set:', unique_nums)
unique_nums.add(6)
group_b = {4, 5, 6, 7}
print('Union:', unique_nums.union(group_b))
print('Intersection:', unique_nums.intersection(group_b))

# Range Operations
r = range(1, 10)
print('Range values:', end=' ')
for val in r:
    print(val, end=' ')

# CONCLUSION
# This practical provided hands-on exposure to all major Python sequence data types —
# strings, lists, arrays, tuples, dictionaries, sets, and range objects. Through indexing, slicing,
# and type-specific operations such as set union/intersection and dictionary keyvalue access,
# the distinct behavior and use cases of each type were clearly demonstrated. Understanding
# when and how to apply each data structure is foundational to writing e icient and readable
# Python programs.
```
