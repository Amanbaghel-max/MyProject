Q1. What is the significance of data visualization in analysis, and how does it support EDA? Data visualization is a critical component of the analytical process because it translates abstract numerical data into visual representations that the human brain can process far more rapidly. Charts, graphs, and plots reveal trends, clusters, correlations, and anomalies that would be diAicult or impossible to detect in raw tabular form. How does Matplotlib diAer from other Python visualization libraries like Seaborn and Plotly? Matplotlib is a foundational, low-level plotting library oAering maximum flexibility and control over every aspect of a figure, at the cost of requiring more verbose code for complex plots. Seaborn is built on top of Matplotlib and provides a high-level API specifically designed for statistical visualization, oAering attractive default themes and built-in support for complex plots like pair plots and violin plots with minimal code. Plotly focuses on interactive, browser-based visualizations suitable for dashboards and web applications. Each library occupies a distinct niche: Matplotlib for fine-grained control, Seaborn for statistical clarity, and Plotly for interactivity. Q3. What is the purpose of the Pyplot module within Matplotlib? The pyplot module provides a state-based, procedural interface to Matplotlib's object oriented plotting system. It was designed to mirror MATLAB's plotting style, allowing users to build figures incrementally with simple function calls rather than managing f igure and axes objects explicitly. Functions such as plt.plot(), plt.scatter(), plt.xlabel(), plt.title(), and plt.show() are all part of the pyplot interface. It handles figure creation, axes management, and rendering, serving as the most accessible entry point for users who want to produce quality visualizations with minimal code. Q4. Describe the steps involved in creating a basic line plot in Matplotlib. Creating a line plot in Matplotlib involves the following sequence: first, import matplotlib.pyplot as plt; second, prepare the x-axis and y-axis data as lists or arrays; third, call plt.plot(x, y) to generate the line; fourth, annotate the chart using plt.title(), plt.xlabel(), and plt.ylabel() for clarity; optionally, enable a grid with plt.grid(True) to improve readability; and finally, display the figure using plt.show(). Each step builds on the previous, and additional customization such as line color, style, and markers can be layered on using keyword arguments to plt.plot(). Q5.
```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

np.random.seed(42)
dataset = pd.DataFrame({
'X': np.arange(1, 101),
'Y': np.random.randint(10, 100, 100),
'Category': np.random.choice(['A', 'B', 'C'], 100)
})
print(dataset.head())

# Scatter Plot
plt.figure(figsize=(7, 4))
plt.scatter(dataset['X'], dataset['Y'], color='steelblue', alpha=0.7)
plt.title('Scatter Plot — X vs Y')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Line Plot
plt.figure(figsize=(7, 4))
plt.plot(dataset['X'], dataset['Y'], color='coral', linewidth=1.2)
plt.title('Line Plot — X vs Y')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()

# Bar Plot
cat_counts = dataset['Category'].value_counts()
plt.figure(figsize=(5, 4))
plt.bar(cat_counts.index, cat_counts.values, color=['#2196F3','#FF5722','#4CAF50'])
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()


# Histogram
plt.figure(figsize=(6, 4))
plt.hist(dataset['Y'], bins=10, color='mediumpurple', edgecolor='white')
plt.title('Histogram of Y Values')
plt.xlabel('Y Values')
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(4, 5))
plt.boxplot(dataset['Y'], patch_artist=True,
            boxprops=dict(facecolor='lightcyan'))
plt.title('Box Plot of Y Values')
plt.ylabel('Y Values')
plt.show()

# Pair Plot
sns.pairplot(dataset)
plt.suptitle('Pair Plot', y=1.02)
plt.show()

# Iris Dataset Visualizations

# Load Iris dataset from sklearn
iris_data = load_iris(as_frame=True)
iris = iris_data.frame
iris.columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']

print(iris.head())
print(iris.describe())
# Iris Scatter — Sepal vs Petal Length
plt.figure(figsize=(7, 4))
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],
color='teal', alpha=0.6)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
# Iris Pair Plot with Species Hue
sns.pairplot(iris, hue='Species')
plt.show()
```
