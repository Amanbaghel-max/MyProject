Theory

Q1. What is web scraping, and what role does the Requests library play in the process? Web scraping is an automated technique for extracting data from websites by programmatically accessing and parsing their HTML content. It enables the collection of large volumes of structured information — such as product prices, news headlines, or book listings — that would be tedious to gather manually. The Requests library in Python acts as the communication layer between the scraper and the web server, issuing HTTP GET or POST requests and retrieving the server's response, which typically contains the page's HTML source. Once this raw HTML is obtained via Requests, it is passed to a parsing library such as BeautifulSoup for data extraction. Q2. What is BeautifulSoup and how does it assist in HTML parsing? BeautifulSoup is a Python library designed to parse and navigate HTML and XML documents. It converts the raw HTML string returned by a web server into a hierarchical tree of Python objects, mirroring the document's DOM structure. This tree representation allows developers to locate, filter, and extract specific elements using tag names, CSS classes, IDs, or attribute values without needing to write complex regular expressions. BeautifulSoup is particularly robust when dealing with malformed or inconsistently structured HTML, as it applies error correction during parsing and normalizes the markup for consistent traversal. Q3. DiAerentiate between web scraping and web crawling. Web scraping and web crawling address diAerent aspects of automated web data collection. Web scraping is focused and targeted — it extracts specific data elements from known pages, such as prices from a product listing or titles from a book catalogue. Web crawling, on the other hand, is exploratory and broad — a crawler systematically follows hyperlinks across many pages to discover and index web content, similar to how search engine bots map the internet. In practice, crawling is often a prerequisite step that identifies the URLs to scrape, while scraping handles the actual data extraction from those discovered pages. Q4. What are the most frequently used methods in BeautifulSoup? BeautifulSoup provides a rich set of navigation and search methods. find() returns the f irst matching element for a given tag or attribute, while find_all() retrieves all matching elements as a list. get_text() extracts the visible text content from an element, stripping all HTML tags. get() retrieves the value of a specific attribute, such as the href of an anchor tag. select() supports CSS selector syntax for more complex targeting. Navigation properties such as .parent, .children, .next_sibling, and .previous_sibling allow traversal through the document tree in relation to a found element. Q5. What ethical and legal considerations apply to web scraping? Responsible web scraping requires awareness of both ethical norms and legal boundaries. The robots.txt file published by most websites specifies which sections may be accessed programmatically and should always be respected before scraping. Excessive or rapid-fire requests can degrade server performance, so introducing time delays between requests is considered good practice. Scraping protected or copyrighted content without explicit permission may violate intellectual property law or the website's terms of service. Additionally, collecting personally identifiable information without user consent raises privacy concerns and may conflict with regulations such as GDPR. Ethical scrapers operate transparently and use data responsibly.

```python
# Implementation
# Installing Required Libraries
!pip install requests beautifulsoup4
# Importing Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
# Sending HTTP Request
target_url = 'http://books.toscrape.com/'
response = requests.get(target_url)
print('HTTP Status Code:', response.status_code)  # 200 = OK

#Parsing HTML Content
soup = BeautifulSoup(response.text, 'html.parser')
print('Page Title:', soup.title.text)

#Extracting Book Titles
book_tags = soup.find_all('h3')
titles = [tag.find('a')['title'] for tag in book_tags]
print('Titles extracted:', len(titles))
print(titles[:5])

#Extracting Prices
price_tags = soup.find_all('p', class_='price_color')
prices = [p.get_text(strip=True) for p in price_tags]
print('Prices extracted:', len(prices))
print(prices[:5])

#Extracting Star Ratings
rating_tags = soup.find_all('p', class_='star-rating')
ratings = [tag['class'][1] for tag in rating_tags]
print('Ratings extracted:', len(ratings))
print(ratings[:5])

#Creating DataFrame
books_df = pd.DataFrame({
'Title' : titles,
'Price' : prices,
'Rating': ratings
})
print(books_df.head(10))
# Saving to CSV
books_df.to_csv('scraped_books.csv', index=False)
print('Data saved to scraped_books.csv')
```
