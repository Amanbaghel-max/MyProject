Theory

Distinguish between Binomial, Poisson, and Normal distributions. • Binomial Distribution applies when the total number of trials is fixed and each trial yields one of two outcomes — success or failure — each with a constant probability. • Poisson Distribution is a discrete probability model suited for counting events that occur randomly within a defined interval of time or space, assuming events are independent of each other. • Normal Distribution is a continuous, bell-shaped, symmetric distribution whose shape is fully described by its mean and standard deviation. In summary: Binomial suits fixed-trial scenarios, Poisson suits event-count scenarios, and Normal suits continuous measurement data.

Practical significance of the Poisson distribution in real life. The Poisson distribution is a useful tool for modeling situations in which we observe how frequently a specific event takes place within a given time period or region. Practical examples include the number of customer service calls arriving per hour, the count of road accidents occurring on a particular stretch in a day, or the number of visitors landing on a website per minute. Its significance lies in enabling organizations to predict demand, allocate resources efficiently, and design systems — such as communication networks, traffic signals, and queuing systems — more effectively.

Central Limit Theorem (CLT) — Concept and Significance. The Central Limit Theorem is a fundamental principle in statistics stating that as the size of a random sample grows larger, the distribution of the sample mean will approximate a normal distribution — regardless of the shape of the original population distribution. This is significant because it justifies the widespread use of normal distribution-based statistical techniques even when the underlying data does not follow a normal pattern, provided the sample size is sufficiently large.

Reasons for the widespread use of Normal distribution. The Normal distribution holds a central position in statistics and data science for several reasons: • A large number of naturally occurring phenomena — such as human heights, exam scores, and measurement errors — tend to follow this distribution. • It is mathematically tractable, making computations of probabilities and statistical tests straightforward. • It forms the theoretical foundation for many inferential statistics methods and machine learning algorithms. Common real-world examples include blood pressure readings, manufacturing tolerances, and standardized test scores.

What is skewness, and how does it relate to each distribution? Skewness quantifies the degree of asymmetry observed in a probability distribution relative to its mean. • Positive skewness (right-skewed): The tail extends further toward higher values. • Negative skewness (left-skewed): The tail extends further toward lower values. • Zero skewness: The distribution is perfectly symmetric about its center. With respect to the three distributions: Binomial can be either skewed or symmetric depending on the probability parameter p; Poisson is generally right-skewed, especially at small values of lambda; and Normal distribution exhibits zero skewness by definition.
